```python
#先導入資料處理會用到的模組
import numpy as np
import numpy.random as random
import scipy as sp
from pandas import Series, DataFrame
import pandas as pd

# 可視化模組
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

# 機器學習模組
import sklearn

# 表示到小數第三位
%precision 3
```

```python
# 開啟 CSV 檔案
dataset= pd.read_csv('data.csv')
```

```python
dataset.head()
```

```python
#資料前處理（逗號轉回小數點）
dataset['% Iron Feed'] = dataset['% Iron Feed'].str.replace(',','.').astype(float)
dataset['% Silica Feed'] = dataset['% Silica Feed'].str.replace(',','.').astype(float)
dataset['Starch Flow'] = dataset['Starch Flow'].str.replace(',','.').astype(float)
dataset['Amina Flow'] = dataset['Amina Flow'].str.replace(',','.').astype(float)
dataset['Ore Pulp Flow'] = dataset['Ore Pulp Flow'].str.replace(',','.').astype(float)
dataset['Ore Pulp pH'] = dataset['Ore Pulp pH'].str.replace(',','.').astype(float)
dataset['Ore Pulp Density'] = dataset['Ore Pulp Density'].str.replace(',','.').astype(float)
dataset['Flotation Column 01 Air Flow'] = dataset['Flotation Column 01 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 02 Air Flow'] = dataset['Flotation Column 02 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 03 Air Flow'] = dataset['Flotation Column 03 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 04 Air Flow'] = dataset['Flotation Column 04 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 05 Air Flow'] = dataset['Flotation Column 05 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 06 Air Flow'] = dataset['Flotation Column 06 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 07 Air Flow'] = dataset['Flotation Column 07 Air Flow'].str.replace(',','.').astype(float)
dataset['Flotation Column 01 Level'] = dataset['Flotation Column 01 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 02 Level'] = dataset['Flotation Column 02 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 03 Level'] = dataset['Flotation Column 03 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 04 Level'] = dataset['Flotation Column 04 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 05 Level'] = dataset['Flotation Column 05 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 06 Level'] = dataset['Flotation Column 06 Level'].str.replace(',','.').astype(float)
dataset['Flotation Column 07 Level'] = dataset['Flotation Column 07 Level'].str.replace(',','.').astype(float)
dataset['% Iron Concentrate'] = dataset['% Iron Concentrate'].str.replace(',','.').astype(float)
dataset['% Silica Concentrate'] = dataset['% Silica Concentrate'].str.replace(',','.').astype(float)
dataset.head()
```

```python
# 為了資料分割(訓練資料與測試資料)的匯入
from sklearn.model_selection import train_test_split
# 為了多元線性迴歸模型建構的導入
from sklearn.linear_model import LinearRegression

 # 指定目標變數為price、其他為解釋變數
X = dataset.drop('% Silica Concentrate', axis=1)
Y = dataset['% Silica Concentrate']

# 分為訓練資料與測試資料
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.8, random_state=0)

# 多元線性迴歸的初始化學習
model = LinearRegression()
model.fit(X_train,Y_train)

# 顯示決定係數
print('決定係數(test):{:.3f}'.format(model.score(X_test,Y_test)))
```
