```python
#導入 matplotlib.pyplot為plt
#從bs4 導入 BeautifulSoup
#導入 requests 模組
import numpy as no
import matplotlib.pyplot as plt
import matplotlib as mpl
from bs4 import BeautifulSoup
import requests
import seaborn as sns

%matplotlib inline

#宣告month, high, low三個list
month,high,low = [],[],[]
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20200101&stockNo=2330'

#使用requests get url
reponse = requests.get(url)

#使用BeautifulSoup解讀
sp = BeautifulSoup(reponse.text, 'html')

#抓出table，並存成變數datas
datas = sp.select("table")[0]

#在datas中選擇tbody的標籤
datas_second = datas.select("tbody")[0]#進入年度資料集

#使用迴圈找出td，讀取出每個月的資料(即tr)
for i in range(12):
    #在datas中選擇tr的標籤
    datas_third = datas_second.select("tr")[i]#每次跳一個月
    #td的第一個欄位為年份，不管他
    #td的第二個欄位為月份，加入到month list中
    month.append(datas_third.select("td")[1].text)
    #td的第三個欄位為當月最高價，加入到high list中
    high.append(datas_third.select("td")[2].text)
    #td的第四個欄位為當月最低價，加入到low list中
    low.append(datas_third.select("td")[3].text)

plt.figure(figsize = (15, 10), dpi = 100, linewidth = 2)

month = [float(i) for i in month]
high = [float(i) for i in high]
low = [float(i) for i in low]

plt.plot(month, high, '-', color = 'blue', label = 'Highest')
plt.plot(month, low, '-', color = 'orange', label = 'Lowest')
plt.legend()

plt.title(2330)
plt.xlabel('month')
plt.ylabel('Money')

#將圖表顯示出來
plt.show()
```
