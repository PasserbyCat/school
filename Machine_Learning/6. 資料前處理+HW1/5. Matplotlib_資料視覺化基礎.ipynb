```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame


import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline


%precision 3
```

# 長條圖

```python
# 顯示的資料
x = [1, 2, 3]
y = [10, 1, 4]

# 指定圖形的大小 
plt.figure(figsize = (10, 6))

# 畫長條圖的各個標籤
plt.bar(x, y, align = 'center', width = 0.5)#底/資料高度/xticks的位置(分edge和center)/bar的寬度

# 長條圖的各個標籤
plt.xticks(x, ['A Class', 'B Class', 'C Class'])

# 設定x與y的標籤
plt.xlabel('Class')
plt.ylabel('Score')

# 顯示格線
plt.grid(True)
```

# 橫條圖

```python
# 顯示的資料
x = [1, 2, 3]
y = [10, 1, 4]

# 指定圖形的大小 
plt.figure(figsize = (10, 6))

#畫橫條圖
plt.barh(x, y, align = 'center')#底/資料高度/xticks的位置(分edge和center)

# 長條圖的各個標籤
plt.yticks(x, ['A Class','B Class','C Class'])

# 設定x與y的標籤
plt.ylabel('Class')
plt.xlabel('Score')

plt.grid(True)
```

# 描繪多個長條圖

```python
#資料的準備
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

# X軸的資料
x = np.arange(len(y1))

# 圖形的寬度
w = 0.4

# 指定圖形的大小
plt.figure(figsize = (10, 6))

# 描繪圖形，對於y2，必須要右移一個寬度
plt.bar(x, y1, color = 'blue', width = w, label = 'Math first', align = 'center')
plt.bar(x + w, y2, color = 'green', width = w, label = 'Math final', align = 'center')
# 讓圖例配置於最適當的位置
plt.legend(loc = 'best')

# x軸要往右移一個寬度
plt.xticks(x + w / 2, ['Class A', 'Class B', 'Class C'])#設定標籤的位置

plt.grid(True)
```

# 堆疊長條圖

```python
# 資料的準備
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

# X軸
x = np.array([1, 2, 3, 4, 5]) 

# 指定圖形大小
plt.figure(figsize = (10, 6))

# 畫堆疊長條圖
p1 = plt.bar(x, height1, color = 'blue')
p2 = plt.bar(x, height2, bottom = height1, color = 'lightblue')#在同樣的x從bottom開始往上畫

# 顯示不同顏色代表什麼
plt.legend((p1[0], p2[0]), ('Class 1', 'Class 2'))
```

# 圓餅圖

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) #第二塊與其他的距離為0.1

# 指定圖形大小
plt.figure(figsize = (15, 6))

# 顯示圖型
plt.pie(sizes, explode = explode, labels = labels, colors = colors,#explode可調整使特定元素脫離群體
       autopct = '%1.1f%%', shadow = False, startangle = 90)#autopct定義各自占比的格式,可定義counterclock(預設為true=順時鐘)

# 維持比例描繪圖形
plt.axis('equal')#確保他式圓形而不是橢圓
```

# 泡泡圖

```python
N = 25

# 隨機生成X,Y資料
x = np.random.rand(N)
y = np.random.rand(N)

# color編號
colors = np.random.rand(N)

# 讓泡泡的大小顯得散亂
area = 10 * np.pi * (15 * np.random.rand(N)) ** 2

plt.figure(figsize = (15, 6))

# 描繪圖形
plt.scatter(x, y, s = area, c = colors, alpha = 0.5)#alpha=明度/s定義範圍/x,y則為原點座標


plt.grid(True)
```
