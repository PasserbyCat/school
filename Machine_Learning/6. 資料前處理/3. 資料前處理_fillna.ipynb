```python
#資料的準備
import numpy as np
from numpy import random
from numpy import nan as NA
import pandas as pd
random.seed(0)
df = pd.DataFrame(np.random.rand(10, 4))
# 設定為NA
df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3] = NA
print(df, '\n\n')
df = df.dropna()#將所有NaN的row刪掉
print(df)
```

```python
#資料的準備
import numpy as np
from numpy import random
from numpy import nan as NA
import pandas as pd
random.seed(0)
df = pd.DataFrame(np.random.rand(10, 4))
# 設定為NA
df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3] = NA
print(df, '\n\n')
df = df.fillna(method = 'ffill')#取前一列的值來填補
print(df)
```

```python
#資料的準備
import numpy as np
from numpy import random
from numpy import nan as NA
import pandas as pd
random.seed(0)
df = pd.DataFrame(np.random.rand(10, 4))
# 設定為NA
df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3] = NA
print(df, '\n\n')
print(df.mean(), '\n\n')
df = df.fillna(df.mean())#取前一列的值來填補
print(df)
```
