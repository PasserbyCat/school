```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

#在Jupyter Notebook上顯示讓圖表直接在該位置
%matplotlib inline
```

# 散佈圖

```python
#散佈圖
import numpy.random as random

#　固定種子值
random.seed(0)

# x軸的資料
x = np.random.randn(30)

# y軸的資料
y = np.sin(x) + np.random.randn(30)

# 指定圖形的大小（可以試著修改20或6)
plt.figure(figsize = (20, 6))

# 描繪圖形
plt.plot(x, y, 'o')#'o'是用來指定描繪的形狀

#也能使用scatter描繪散佈圖
#plt.scatter(x, y)

# 標題
plt.title('Title Name')
# X軸標籤
plt.xlabel('X')
# Y軸標籤
plt.ylabel('Y')

# 顯示grid（圖形中的格線）
plt.grid(True)
```

# 連續值散佈圖

```python
# 固定種子值
np.random.seed(0)

# 資料範圍
numpy_data_x = np.arange(1000)

# 生成亂數與累積和
numpy_random_data_y = np.random.randn(1000).cumsum()

# 指定圖形大小
plt.figure(figsize=(20, 6))

# 可藉由label= 與 legand來加上標籤
plt.plot(numpy_data_x, numpy_random_data_y, label='Label')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

 # 圖形分割

```python
plt.figure(figsize=(20, 6))

#  2列1行的圖形第一個
plt.subplot(2, 1, 1)
# 從-10~10分割出100個數字之串列並取出
x = np.linspace(-10, 10, 100)
plt.plot(x, np.sin(x))

# 2列1行的圖形第二個
plt.subplot(2, 1, 2)
y = np.linspace(-10, 10,100)
plt.plot(y, np.sin(2 * y))

plt.grid(True)
```

# 畫出函數圖形

```python
def my_function(x):
    return x ** 2 + 2 * x + 1

x = np.arange(-10, 10)
plt.figure(figsize = (20, 6))

#畫出函數圖形
plt.plot(x, my_function(x))

plt.grid(True)
```

# 直方圖

```python
random.seed(0)

# 指定圖形的大小
plt.figure(figsize = (20, 6))

# 畫直方圖
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins = 60, range = (20, 80))#資料/個數/範圍

plt.grid(True)
```

# 動手做

```python
plt.figure(figsize=(20, 6))

#  2列1行的圖形第一個
plt.subplot(2, 1, 1)
# 從-10~10分割出100個數字之串列並取出
plt.hist(np.random.randn(1000), bins = 10, range = (0.0, 1.0))#資料/個數/範圍

# 2列1行的圖形第二個
plt.subplot(2, 1, 2)
plt.hist(np.random.randn(1000), bins = 10, range = (0.0, 1.0))#資料/個數/範圍

plt.grid(True)
```
